```python
import bibtexparser
with open('export.bib') as f:
    bib = bibtexparser.load(f)
```

```python
selected_bib = []
for entry in bib.entries:
    if entry['ENTRYTYPE'] in ['inproceedings', 'article']:
        selected_bib.append(entry)
```

```python
import pandas as pd
import os
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time

if not os.path.exists('./outputs'):
    os.mkdir('./outputs')
driver_path  = './chromedriver_win32/chromedriver.exe'
url_dir = 'https://search.crossref.org/?from_ui=yes&q='
driver = webdriver.Chrome(driver_path)

columns = ['org_title','ref_title','doi']
df = pd.DataFrame([],columns = columns) # review [ 'sadfasdf', 'asfasdf',...]
```

```python
for entry in selected_bib:
    url = url_dir + entry['title']
    
    driver.get(url)
    try:
        ref_title_tag = driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[2]/table/tbody/tr[1]/td/p[1]')
        ref_doi_tag   = driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[2]/table/tbody/tr[1]/td/div/div/a')

        ref_title     = ref_title_tag.text.strip()
        ref_doi       = ref_doi_tag.text.strip()
        ref_doi       = ref_doi.replace('https://doi.org/','')
    except: 
        ref_title = 'error'
        ref_doi = None
    
    print([entry['title'],ref_title,ref_doi])
    df = df.append(dict(zip(columns,[entry['title'],ref_title,ref_doi])) ,ignore_index=True)
```

```python
df.to_csv('outputs/doi_list.csv')
```
